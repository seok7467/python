```python
import time
t1 = time.time()

# 모든 정보를 담을 DF
movieDOF = pd.DataFrame()

# 1개 영화 정보들에 대해서 처리할 DF
movieDOF_temp = pd.DataFrame()
#movieDOF_temp = movieDOF_temp.append({"movieCd":"","movieNm":"","movieNmEn":"","openDt":"","directorpeopleNm":""},ignore_index=True)

num=len(movieData["movieListResult"]["movieList"])

for i in range(0,num):
  
    if movieData["movieListResult"]["movieList"][i]["directors"]:
        a=movieData["movieListResult"]["movieList"][i]["directors"][0]["peopleNm"]
        movieDOF_temp = movieDOF_temp.append({"movieCd":movieData["movieListResult"]["movieList"][i]["movieCd"],
                                          "movieNm":movieData["movieListResult"]["movieList"][i]["movieNm"],
                                          "movieNmEn":movieData["movieListResult"]["movieList"][i]["movieNmEn"],
                                          "openDt":movieData["movieListResult"]["movieList"][i]["openDt"],
                                          "directorpeopleNm":a
                                         },ignore_index=True)
    else:
        
        movieDOF_temp = movieDOF_temp.append({"movieCd":movieData["movieListResult"]["movieList"][i]["movieCd"],
                                          "movieNm":movieData["movieListResult"]["movieList"][i]["movieNm"],
                                          "movieNmEn":movieData["movieListResult"]["movieList"][i]["movieNmEn"],
                                          "openDt":movieData["movieListResult"]["movieList"][i]["openDt"],
                                          "directorpeopleNm":""
                                         },ignore_index=True)
    
    
    # 1. movieDOF_temp에 i번째 영화들의 정보 담기(코드, 이름, 영문, 개봉일, 감독)
    # 2. movieDOF_temp에 있는 i번째 영화에 대한 정보를 movieDOF에 추가
    # append활용
    # 3. movieDOF_temp에 담겨져 있는 정보들을 초기화 하는 부분
t2 = time.time()
print ("방법1 : " + str(t2-t1))
movieDOF_temp
```

```python
import time
t1 = time.time()

movieDOF = pd.DataFrame()


movieDOF_temp = pd.DataFrame()

movieDOF_temp = movieDOF_temp.append({"movieCd":"","movieNm":"","movieNmEn":"","openDt":"","directorpeopleNm":""},ignore_index=True)


temp_movcd_idx=movieDOF_temp.columns.get_loc("movieCd")
temp_movNm_idx=movieDOF_temp.columns.get_loc("movieNm")
temp_movNmEn_idx=movieDOF_temp.columns.get_loc("movieNmEn")
temp_movdt_idx=movieDOF_temp.columns.get_loc("openDt")
temp_movdir_idx=movieDOF_temp.columns.get_loc("directorpeopleNm")

num=len(movieData["movieListResult"]["movieList"])


for i in range(0,num):
    movieDOF_temp.iat[0,temp_movcd_idx]=movieData["movieListResult"]["movieList"][i]["movieCd"]
    
    movieDOF_temp.iat[0,temp_movNm_idx]=movieData["movieListResult"]["movieList"][i]["movieNm"]
    
    movieDOF_temp.iat[0,temp_movNmEn_idx]=movieData["movieListResult"]["movieList"][i]["movieCd"]
    
    movieDOF_temp.iat[0,temp_movdt_idx]=movieData["movieListResult"]["movieList"][i]["openDt"]
    
    if movieData["movieListResult"]["movieList"][i]["directors"] !=[]:
        movieDOF_temp.iat[0,temp_movdir_idx]=movieData["movieListResult"]["movieList"][i]["directors"][0]["peopleNm"]
    else:
        pass
    
    movieDOF=movieDOF.append(movieDOF_temp, ignore_index=True)
    
    for j in range(0,len(movieDOF_temp.columns)):
        movieDOF_temp.iat[0,j]=""

        
        



        
t2 = time.time()
print ("방법2 : " + str(t2-t1))
```

```python
# 컬럼 추가하기

"""
year = movieDOF["openDt"][0][:4]
month= movieDOF["openDt"][0][4:6]
month= movieDOF["openDt"][0][6:]

"""

movieDOF["year"] = movieDOF["openDt"].apply(lambda x: str(x)[:4])
movieDOF["month"] = movieDOF["openDt"].apply(lambda x: str(x)[4:6])
movieDOF["day"] = movieDOF["openDt"].apply(lambda x: str(x)[6:])
movieDOF
```

```python
# 정렬하기

movieDOF=movieDOF.sort_values(by="openDt")
#movieDOF=movieDOF.sort_index()
movieDOF
```
