# pivot

```python
import pandas as pd
import numpy as np
```

```python
### Pivot ###
# ref) http://pbpython.com/pandas-pivot-table-explained.html
# data ) sales-funnel.xlsx
data = pd.read_excel("sampledata/11_sales-funnel.xlsx")
data.head()
```

```python
data.head()
```

```python
# 좀 더 명확하게 하기위해서 카테고리 변수화
data["Status"] = data["Status"].astype("category")

data["Status"].cat.set_categories(
    ["won","pending","presented","declined"],
    inplace=True)

data.head()
```

```python
data.info()
```

```python
# Name을 가로줄 기준으로 하고, 데이터를 보자 --> 숫자인 컬럼들만 들어온다!
pd.pivot_table(data,index=["Name"])
```

```python
# 가로줄 여러 계층 구조로 사용할 경우
pd.pivot_table(data,index=["Name","Rep","Manager"])
```

```python
# 2개만 가로줄 선택
pd.pivot_table(data,index=["Manager","Rep"])
```

```python
# row - manager / rep, column -= price
pd.pivot_table(data,index=["Manager","Rep"],
               values=["Price"])
```

```python
# pivot with aggfunc : sum
import numpy as np
pd.pivot_table(data,index=["Manager","Rep"],
               values=["Price"],
               aggfunc=np.sum)
```

```python
# pivot with aggfunc --> make new column
pd.pivot_table(data,index=["Manager","Rep"],
               values=["Price"],
               aggfunc=[np.mean,len])
```

```python
# 세로줄에 대한 계층
pd.pivot_table(data,index=["Manager","Rep"],
               values=["Price"], 
               columns=["Product"],
               aggfunc=[np.sum])
```

```python
# 세로줄에 대한 계층- fillvalue
pd.pivot_table(data,
               index=["Manager","Rep"],
               values=["Price"], 
               columns=["Product"],
               aggfunc=[np.sum], 
               fill_value=0)
```

```python
# 여러 컬럼에 여러 계층 구조!
pd.pivot_table(data,index=["Manager","Rep"],
               values=["Price","Quantity"], 
               columns=["Product"],
               aggfunc=[np.sum],
               fill_value=0)
```

```python
# 품목에 대한 정보를 가로 인덱스로 
pd.pivot_table(data,index=["Manager","Rep","Product"],
               values=["Price","Quantity"],
               aggfunc=[np.sum],
               fill_value=0)
```

```python
# Total Option
pd.pivot_table(data,index=["Manager","Rep","Product"], 
               values=["Price","Quantity"],
               aggfunc=[np.sum,np.mean],
               fill_value=0,
               margins=True)
```

```python
# row : manager/status 
# column : Quantity [ len ]/ Price [ sum]
# column category : product
pd.pivot_table(data,index=["Manager","Status"],
            columns=["Product"],
            values=["Quantity","Price"],
        aggfunc={"Quantity":len,"Price":np.sum},
            fill_value=0)
```

```python
# aggfunc  에서도 계층 구조 적용
pd.pivot_table(data,index=["Manager","Status"],
            columns=["Product"],
            values=["Quantity","Price"], 
aggfunc={"Quantity":len,"Price":[np.sum,np.mean]},
               fill_value=0)
```

```python
# pivot & filter
table = pd.pivot_table(data,index=["Manager","Status"],columns=["Product"],values=["Quantity","Price"], 
               aggfunc={"Quantity":len,"Price":[np.sum,np.mean]},fill_value=0)
table.query('Manager == ["Debra Henley"]')
```

```python
table.query('Status == ["pending","won"]')
#?table.query(Status == ["pending","won"])
```

```python
### crosstab ###
pd.crosstab(data["Name"], data["Price"])
```

```python
pd.crosstab([ data["Manager"], data["Rep"]  ], [  ])
```

```python
pd.crosstab([data["Manager"], data["Rep"]], 
            [data["Price"], data["Quantity"]])
```

```python
pd.crosstab([data["Manager"], data["Rep"]], 
            [data["Price"], data["Quantity"]], 
            margins=True)
```

```python
pd.crosstab([data["Manager"], data["Rep"]], 
            data["Product"], 
            values=data["Price"], 
            aggfunc=[len, np.mean], 
            margins=True)
```

```python
pd.crosstab([data["Manager"], data["Rep"]], 
            data["Product"],
            values=data["Price"], 
            aggfunc=[len, np.mean], 
            margins=True).fillna(0)
```

```python
#######  stack / unstack  참고!!!!####
# Row Multi-Index
row_idx_arr = list(zip(['r0', 'r0'], ['r-00', 'r-01']))
row_idx = pd.MultiIndex.from_tuples(row_idx_arr)
 
# Column Multi-Index
col_idx_arr = list(zip(['c0', 'c0', 'c1'], ['c-00', 'c-01', 'c-10']))
col_idx = pd.MultiIndex.from_tuples(col_idx_arr)
 
# Create the DataFrame
d = pd.DataFrame(np.arange(6).reshape(2,3), index=row_idx, columns=col_idx)
d = d.applymap(lambda x: (x // 3, x % 3))
 
# Stack/Unstack
s = d.stack()
u = d.unstack()
```
