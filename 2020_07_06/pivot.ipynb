```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_excel("sampledata/11_sales-funnel.xlsx")
data.info()
```

```python
data["Status"] = data["Status"].astype("category")

data["Status"].cat.set_categories(
    ["won","pending","presented","declined"],
    inplace=True)

data.info()   # object가 category로 바뀜
data
```

```python
pd.pivot_table(data,index=["Name"]) # 숫자 컬럼들만 들어오네, #같은 이름이면 다 더한 후 평균값을 가져옴
```

```python
pd.pivot_table(data,index=["Name","Rep","Manager"])
```

```python
pd.pivot_table(data,index=["Manager","Rep"])
pd.pivot_table(data,index=["Manager","Rep"],values=["Price"])
pd.pivot_table(data,index=["Manager","Rep"],values=["Price"],aggfunc=np.sum) # 총 매출액(sum)
pd.pivot_table(data,index=["Manager","Rep"],values=["Price"],aggfunc=[np.mean,len]) # 평균과 매출 횟수
pd.pivot_table(data,index=["Manager","Rep"],values=["Price"],aggfunc=[np.sum],columns=["Product"]) # 상품별 표시
pd.pivot_table(data,index=["Manager","Rep"],values=["Price","Quantity"],aggfunc=[np.sum],columns=["Product"],fill_value=0) #NAN채우기 

pd.pivot_table(data,index=["Manager","Rep"],
               values=["Price","Quantity"],
               aggfunc=[np.sum,np.mean],
               columns=["Product"],
               fill_value=0,
               margins=True)  #총합

table = pd.pivot_table(data,index=["Manager","Status"],
               values=["Price","Quantity"],
               aggfunc={"Price":np.sum,"Quantity":len}, #딕셔너리로 해서 표시해봐
               columns=["Product"],
               fill_value=0,
               margins=True)  #총합

table
```

```python
# table.query('Status==["won"]')
table.query('Status==["won","pending"]')
```

```python
pd.crosstab(data["Name"],data["Price"])
pd.crosstab([data["Manager"],data["Rep"]],[data["Price"],data["Quantity"]])
pd.crosstab([data["Manager"],
             data["Rep"]],
            data["Product"],
            values=data["Price"],
            aggfunc=[len,np.mean]).fillna(0)
             
```
