* International Plan의 값을 확인을 해보니 No/ Yes로 되어 있음. 이러한 것은 앞으로 데이터 처리하기에 불편한 부분이 있기에 No는 False, Yes는 True 인 불리언 값들로 변경!!!

```python
"""
lambda함수 조건문 사용하기
lambda x : [true값] if [조건] else [false값]
lambda x : [true값] if [조건] else( [true값] if [조건] else [false값])

"""

# df["International plan"].apply(lambda x : True if x=="Yes" else  False)

df["International plan"]=df["International plan"].map({"No":False,"Yes":True})
df
```

* 위의 작업을 Voice Mail Plan 컬럼에도 적용해보자!!!

```python
df["Voice mail plan"]=df["Voice mail plan"].map({"No":False,"Yes":True})
df
```

```python
df = df.replace({'Voice mail plan': {"No":False,"Yes":True}})
df.head()
```

* 우선 보고 싶은 컬럼들을 추려서(Total data minutes, Total eve minutes, Total night minutes), 이것들을 Churn의 값을 기준으로 기본적인 정보들(describe)을 보자!!!

```python
cols=["Total day minutes","Total eve minutes","Total night minutes"]
df.groupby(["Churn"])[cols].describe()
```

* 우선 보고 싶은 컬럼들을 추려서(Total data minutes, Total eve minutes, Total night minutes), 이것들을 Churn의 값을 기준으로 평균/표준편차/최소값/최대값만 보자!!

```python
cols=["Total day minutes","Total eve minutes","Total night minutes"]
df.groupby(["Churn"])[cols].agg([np.mean, np.std, np.min, np.max])
```

* 아래와 같은 테이블을 pivot_table을 이용해서 생성하시오, 칸은 평균으로 채울 것!

* total_calls 에는 total day/eve/noght/intl calls을 정리하고, 원본 데이터에다가 total_calls의 값을 추가하시오!!!

```python
total_calls=df["Total day calls"]+df["Total eve calls"]+df["Total night calls"]+df["Total intl calls"]
df.insert(loc=len(df.columns), column='Total calls', value=total_calls) 
df
```

* 위와 비슷한 기능인데, 좀 더 단순하게 Total Charge라는 컬럼으로 생성하시오!!

```python
df["total_charge"]=df["Total day calls"]+df["Total eve calls"]+df["Total night calls"]+df["Total intl calls"]
df
```

* 방금 위에서 생성한 2개의 컬름 total charge, total calls 의 컬럼을 제거하시오!!!!!

```python
df.drop(["total_charge","Total calls"],axis=1,inplace=True)
df
```

* Many Service Calls 라는 컬럼을 만들고 나서, 이 컬럼에는 Customer service calls의 값이 3 초과인 값들을 가지고 처리.  그리고 Many Service calls 라는 값과 churn의 값을 비교하는 crosstab을 만들기!

```python
df["Many_Service_calls"]=(df["Customer service calls"]>3).astype("int")
pd.crosstab(df["Many_Service_calls"],df["Churn"],margins=True)
```

```python
df["cross_manycall_interplan"]= df["Many_Service_calls"] & df["International plan"]
pd.crosstab(df["cross_manycall_interplan"],df["Churn"])
```
