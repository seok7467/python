```python
import time
t1 = time.time()

testCSV_path="sampledata\\01_olive.csv"
olive_oil=pd.read_csv(testCSV_path)
olive_oil.rename(columns={olive_oil.columns[0]:"id_area"}, inplace=True)
for i in range(0,len(olive_oil)):
    olive_oil.iloc[i,0]=str(int(olive_oil.iloc[i,0].split(".")[0])-1) + "_" +olive_oil.iloc[i,0].split(".")[1]

t2 = time.time()
print(olive_oil)
print ("방법1 : " + str(t2-t1))
```

```python
import time
t1 = time.time()

testCSV_path="sampledata\\01_olive.csv"
olive_oil=pd.read_csv(testCSV_path)
olive_oil.rename(columns={olive_oil.columns[0]:"id_area"}, inplace=True)
olive_oil["id_area"]=olive_oil["id_area"].apply(lambda x :str(int(x.split(".")[0])-1) + "_" +x.split(".")[1])

t2 = time.time()
print(olive_oil)
print ("방법1 : " + str(t2-t1))
```

### vars() 문자열을 변수명으로 바꾸기

```python
glob.glob(many_file_path)[0]
```

```python
os.path.basename(glob.glob(many_file_path)[0]).split(".")[0] # 앞에 경로들 다 지워줌.앞에 이름만 가져오기
```

```python
file_path_list=glob.glob(many_file_path)

for i in range(0,len(file_path_list)):
    temp=os.path.basename(file_path_list[i])
    temp=os.path.splitext(temp)[0]           #파일명만 들어내기 (os.path.splitext)
    vars()['data_'+str(temp)]=pd.read_csv(file_path_list[i]) # vars() 사용해서 변수명으로 만들기
```
