```python
import pandas as pd
import urllib.request
from bs4 import BeautifulSoup

sampleURL="http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.xml?key=105d650c6cc06c74308970b42deeca5a&openStartDt=2020&itemPerPage=40"
res=urllib.request.urlopen(sampleURL)
soup=BeautifulSoup(res,"html.parser")

movieDOF = pd.DataFrame()
movieDOF_temp = pd.DataFrame()
movieDOF_temp = movieDOF_temp.append({"movieCd":"","movieNm":"","movieNmEn":"","openDt":"","directorpeopleNm":""},ignore_index=True)

temp_movcd_idx=movieDOF_temp.columns.get_loc("movieCd")
temp_movNm_idx=movieDOF_temp.columns.get_loc("movieNm")
temp_movNmEn_idx=movieDOF_temp.columns.get_loc("movieNmEn")
temp_movdt_idx=movieDOF_temp.columns.get_loc("openDt")
temp_movdir_idx=movieDOF_temp.columns.get_loc("directorpeopleNm")

num=len(soup.find_all("movie"))


for i in range(0,num):
    tempInfo=soup.find_all("movie")[i]
    
    movieDOF_temp.iat[0,temp_movcd_idx]=tempInfo.find("moviecd").text    
    movieDOF_temp.iat[0,temp_movNm_idx]=tempInfo.find("movienm").text
    movieDOF_temp.iat[0,temp_movNmEn_idx]=tempInfo.find("movienmen").text
    movieDOF_temp.iat[0,temp_movdt_idx]=tempInfo.find("opendt").text
    
    
    
    
    if tempInfo.find("directors").text:
        movieDOF_temp.iat[0,temp_movdir_idx]=tempInfo.find("peoplenm").text
    else:
        pass
    
    
    movieDOF=movieDOF.append(movieDOF_temp, ignore_index=True)
    
    for j in range(0,len(movieDOF_temp.columns)):
        movieDOF_temp.iat[0,j]=""

        
movieDOF["year"] = movieDOF["openDt"].apply(lambda x: str(x)[:4])
movieDOF["month"] = movieDOF["openDt"].apply(lambda x: str(x)[4:6])
movieDOF["day"] = movieDOF["openDt"].apply(lambda x: str(x)[6:])

movieDOF=movieDOF.sort_values(by="openDt")
movieDOF=movieDOF.set_index(movieDOF.columns[1])

col_list=["openDt","year","month","day","movieNm","movieNmEn","directorpeopleNm"]
movieDOF = movieDOF.loc[:,col_list]
movieDOF
        
```

```python
movieDOF["typeNm"]=""
movieDOF["peopleNm"]=""
movieDOF["cast"]=""
movieDOF["actors"]=""

for i in range(0,len(movieDOF)):
    tempURL="http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieInfo.xml?"+\
    "key=105d650c6cc06c74308970b42deeca5a&movieCd="+movieDOF.index[i]
    res=urllib.request.urlopen(tempURL)
    soup=BeautifulSoup(res,"html.parser")
    
    # tempMoviePage=urllib.request.urlopen(tempURL)
    # tempMovieData=json.loads(tempMoviePage.read().decode("utf8"),encoding="utf8")
    
   # print(soup.find("movieinforesult").find("movieinfo").find("typenm").text)
    #print(soup.find("movieinforesult").find("movieinfo").find("actors").text)
    
   # typeNm=tempMovieData["movieInfoResult"]["movieInfo"]["typeNm"]
    #peopleNm=tempMovieData["movieInfoResult"]["movieInfo"]["actors"]
   # cast=tempMovieData["movieInfoResult"]["movieInfo"]["actors"]
    #actors=tempMovieData["movieInfoResult"]["movieInfo"]["actors"]

    tempInfo=soup.find("movieinforesult").find("movieinfo")
    if tempInfo.find("typenm").text:
        movieDOF.at[movieDOF.index[i],"typeNm"]=tempInfo.find("typenm").text
    else:
        pass
    if tempInfo.find("actors").text:
        movieDOF.at[movieDOF.index[i],"peopleNm"]=tempInfo.find("actors").find("actor").find("peoplenm").text
    else:
        pass
    if tempInfo.find("actors").text:
        movieDOF.at[movieDOF.index[i],"cast"]=tempInfo.find("actors").find("actor").find("cast").text
    else:
        pass
    if tempInfo.find("actors").text:
        movieDOF.at[movieDOF.index[i],"actors"]=len(tempInfo.find("actors").find("actor").text)
    else:
        pass 
   
movieDOF
```
