```python
import pandas as pd
import urllib.request
import json

movieListURL="http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.json?key=105d650c6cc06c74308970b42deeca5a&itemPerPage=100"
moviePage=urllib.request.urlopen(movieListURL)
movieData=json.loads(moviePage.read().decode("utf8"),encoding="utf8")


movieDOF = pd.DataFrame()
movieDOF_temp = pd.DataFrame()
movieDOF_temp = movieDOF_temp.append({"movieCd":"","movieNm":"","movieNmEn":"","openDt":"","directorpeopleNm":""},ignore_index=True)


temp_movcd_idx=movieDOF_temp.columns.get_loc("movieCd")
temp_movNm_idx=movieDOF_temp.columns.get_loc("movieNm")
temp_movNmEn_idx=movieDOF_temp.columns.get_loc("movieNmEn")
temp_movdt_idx=movieDOF_temp.columns.get_loc("openDt")
temp_movdir_idx=movieDOF_temp.columns.get_loc("directorpeopleNm")

num=len(movieData["movieListResult"]["movieList"])


for i in range(0,num):
    movieDOF_temp.iat[0,temp_movcd_idx]=movieData["movieListResult"]["movieList"][i]["movieCd"]    
    movieDOF_temp.iat[0,temp_movNm_idx]=movieData["movieListResult"]["movieList"][i]["movieNm"]
    movieDOF_temp.iat[0,temp_movNmEn_idx]=movieData["movieListResult"]["movieList"][i]["movieCd"]
    movieDOF_temp.iat[0,temp_movdt_idx]=movieData["movieListResult"]["movieList"][i]["openDt"]
    if movieData["movieListResult"]["movieList"][i]["directors"] !=[]:
        movieDOF_temp.iat[0,temp_movdir_idx]=movieData["movieListResult"]["movieList"][i]["directors"][0]["peopleNm"]
    else:
        pass
    
    movieDOF=movieDOF.append(movieDOF_temp, ignore_index=True)
    
    for j in range(0,len(movieDOF_temp.columns)):
        movieDOF_temp.iat[0,j]=""
        

        
# 컬럼 추가하기
movieDOF["year"] = movieDOF["openDt"].apply(lambda x: str(x)[:4])
movieDOF["month"] = movieDOF["openDt"].apply(lambda x: str(x)[4:6])
movieDOF["day"] = movieDOF["openDt"].apply(lambda x: str(x)[6:])

# 정렬하기
movieDOF=movieDOF.sort_values(by="openDt")
movieDOF=movieDOF.set_index(movieDOF.columns[1])


col_list=["openDt","year","month","day","movieNm","movieNmEn","directorpeopleNm"]
movieDOF = movieDOF.loc[:,col_list]
movieDOF
```

```python
import time
t1 = time.time()
movieDOF["typeNm"]=""
movieDOF["peopleNm"]=""
movieDOF["cast"]=""
movieDOF["actors"]=""
T=[]
P=[]
C=[]
A=[]
for i in range(0,len(movieDOF)):
    
    
    tempURL="http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieInfo.json?key=105d650c6cc06c74308970b42deeca5a&movieCd="+movieDOF.index[i]
    tempMoviePage=urllib.request.urlopen(tempURL)
    tempMovieData=json.loads(tempMoviePage.read().decode("utf8"),encoding="utf8")
    if tempMovieData["movieInfoResult"]["movieInfo"]["typeNm"]:
        T.append(tempMovieData["movieInfoResult"]["movieInfo"]["typeNm"])
    else:
        T.append("")
    
    if tempMovieData["movieInfoResult"]["movieInfo"]["actors"]:
        P.append(tempMovieData["movieInfoResult"]["movieInfo"]["actors"][0]["peopleNm"])
    else:
        P.append("")
    if tempMovieData["movieInfoResult"]["movieInfo"]["actors"]:
        C.append(tempMovieData["movieInfoResult"]["movieInfo"]["actors"][0]["cast"])
    else:
        C.append("")
    
    if tempMovieData["movieInfoResult"]["movieInfo"]["actors"]:
        A.append(len(tempMovieData["movieInfoResult"]["movieInfo"]["actors"]))
    else:
        A.append("")
movieDOF["typeNm"]=T
movieDOF["peopleNm"]=P
movieDOF["cast"]=C
movieDOF["actors"]=A
t2 = time.time()
print ("방법2 : " + str(t2-t1))
```
