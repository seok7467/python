```python
# dataSet
baseball_player=pd.DataFrame(
{
    "player_Name":
    ["권혁","이범호","필","박병호","양준혁"],
    "team_No":
    [1993,2001,2001,2010,None]
}
)
baseball_team=pd.DataFrame(
{
    "team_Name": ["한화","두산","기아","넥센"],
    "team_home": ["대전","서울","광주","고척"],
    "team_id": [1993,1999,2001,2010],
    
}
)

pd.merge(baseball_player,baseball_team, 
         how="inner",                      # inner 방법
         left_on="team_No",                # 왼쪽은 team_No 기준, 기준이 컬럼에 있을 때
         right_on="team_id")               # 오른쪽은 team_id 기준
```

```python
pd.merge(baseball_player,baseball_team,
        how="left",              # 왼쪽은 무조건 살려~
        left_on="team_No",
        right_on="team_id")
```

```python
pd.merge(baseball_player,baseball_team,
        how="right",              # 오른쪽 무조건 살려~
        left_on="team_No",
        right_on="team_id")
```

```python
# left 에만 있는 원소 뽑아내기

temp=pd.merge(baseball_player,baseball_team,
        how="left",              # 왼쪽은 무조건 살려~
        left_on="team_No",
        right_on="team_id")


temp[temp["team_id"].isnull()]
```

```python
# right 에만 있는 원소 뽑아내기

temp=pd.merge(baseball_player,baseball_team,
        how="right",              # 왼쪽은 무조건 살려~
        left_on="team_No",
        right_on="team_id")


temp[temp["team_No"].isnull()]
```

```python
# union 빠진 거 없이 그냥 다모여!~~~~~~~~~~~~~~~~~
pd.merge(baseball_player,baseball_team,
        how="outer",                    # 하나로 합체. 
        left_on="team_No",
        right_on="team_id")
```

```python
# 빠진 거 만 추출해봐 왼쪽이나 오른쪽 상관없이

temp=pd.merge(baseball_player,baseball_team,
        how="outer",              # 왼쪽은 무조건 살려~
        left_on="team_No",
        right_on="team_id")


temp[temp["team_No"].isnull() | temp["team_id"].isnull()]
```

```python
baseball_player=pd.DataFrame(
{
    "player_Name":
    ["권혁","이범호","필","박병호","양준혁"],
    "team_No":
    [1993,2001,2001,2010,None]
},
    index=["p_01","p_02","p_03","p_04","p_05"]
)

baseball_player.index.name="player_id"

# player는 기준이 컬럼: temp_No. 팀 정보는 기준이 가로 인덱스에 존재

baseball_team=pd.DataFrame(
{
    "team_Name": ["한화","두산","기아","넥센"],
    "team_home": ["대전","서울","광주","고척"],
    
    
},
    index=[1993,1999,2001,2010]
)


# inner
pd.merge(baseball_player,baseball_team,
        how="inner",              
        left_on="team_No",
        right_index=True)
```

```python
# left join :선수들 정보는 다 있는 상태에서 팀 정보 가져다 붙이기
# right join
```

```python
pd.merge(baseball_player,baseball_team,
        how="left",                      # 왼쪽 모두 살려~
        left_on="team_No",
        right_index=True)
```

```python
pd.merge(baseball_player,baseball_team,
        how="right",                      # 오룬쪽 모두 살려~
        left_on="team_No",
        right_index=True)
```

```python
# left에만 있는 것 출력
temp=pd.merge(baseball_player,baseball_team,
        how="left",                      # 왼쪽 모두 살려~
        left_on="team_No",
        right_index=True)
temp[temp["team_No"].isnull()]
```

```python
# right 만 있는 것 출력
temp=pd.merge(baseball_player,baseball_team,
        how="right",                      # 오른쪽 모두 살려~
        left_on="team_No",
        right_index=True)

temp[temp.index.isnull()]
```
